```python
import pandas as pd
import os
```

```python
# Load the datasets
training_contextual_path = '/content/drive/Shareddrives/HLT/dataset/development/training_contextual.csv'
training_textual_path = '/content/drive/Shareddrives/HLT/dataset/development/training_textual.csv'
training_contextual_retweets_path = '/content/drive/Shareddrives/HLT/dataset/development/training_contextual_retweets.csv'
training_contextual_replies_path = '/content/drive/Shareddrives/HLT/dataset/development/training_contextual_replies.csv'
training_contextual_friends_path = '/content/drive/Shareddrives/HLT/dataset/development/training_contextual_friends.csv'

# Read the datasets into Pandas DataFrames
df_training_contextual = pd.read_csv(training_contextual_path)
df_training_textual = pd.read_csv(training_textual_path)
df_training_contextual_retweets = pd.read_csv(training_contextual_retweets_path)
df_training_contextual_replies = pd.read_csv(training_contextual_replies_path)
df_training_contextual_friends = pd.read_csv(training_contextual_friends_path)

(df_training_contextual.columns, df_training_textual.columns,
 df_training_contextual_retweets.columns, df_training_contextual_replies.columns,
 df_training_contextual_friends.columns)
```

### Merging training

```python
new_training = df_training_textual.merge(df_training_contextual, on=['anonymized_tweet_id', 'dataset'])
new_training.drop(['source','user_created_at','created_at','is_reply','is_retweet','is_quote'], axis=1, inplace=True)
new_training.info()
```

```python
new_training.dropna(inplace=True)
new_training.reset_index(drop=True, inplace=True)
new_training.info()
```

```python
new_training
```

```python
new_training.to_csv('/content/drive/Shareddrives/HLT/dataset/development/merged_training.csv', index=False)
```
